```python
import pandas as pd
```

```python
matches = pd.read_csv("E:\\2020 WPL WIDHYA\\Day 10\\matches.csv")
deliveries = pd.read_csv("E:\\2020 WPL WIDHYA\\Day 10\\deliveries.csv")
```

```python
matches=matches.replace(['Delhi Daredevils'],'Delhi Capitals')
```

```python
deliveries=deliveries.replace(['Delhi Daredevils'], 'Delhi Capitals')
```

```python
filter1=matches[((matches.winner=='Delhi Capitals')|(matches.winner=='Kings XI Punjab'))
               &
               ((matches.season==2019)|(matches.season==2018)|(matches.season==2017)|(matches.season==2016))]
filter1
```

```python
filter2=deliveries[((deliveries.batting_team=='Delhi Capitals')&(deliveries.bowling_team=='Kings XI Punjab'))
                  |
                  ((deliveries.batting_team=='Kings XI Punjab')&(deliveries.bowling_team=='Delhi Capitals'))]
```

```python
filter2
```

```python
details=pd.merge(filter1, filter2, left_on='id', right_on='match_id')
details.to_csv('details.csv')
```
