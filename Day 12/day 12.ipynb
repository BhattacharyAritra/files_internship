```python
import pandas as pd
```

```python
delivery = pd.read_csv("E:\\2020 WPL WIDHYA\\Day 12\\deliveries.csv")
match = pd.read_csv("E:\\2020 WPL WIDHYA\\Day 12\\matches.csv")
```

```python
#Filtering deliveries dataset by batting and bowling team
fl_dlvr = delivery[((delivery.batting_team=="Rajasthan Royals")&(delivery.bowling_team=="Sunrisers Hyderabad"))
                  |
                  ((delivery.batting_team=="Sunrisers Hyderabad")&(delivery.bowling_team=="Rajasthan Royals"))]
```

```python
#Filtering matches dataset by match winners and seasons
fl_mtch = match[((match.winner=="Rajasthan Royals")|(match.winner=="Sunrisers Hyderabad"))
               &
               ((match.season==2019)|(match.season==2018)|(match.season==2014)|(match.season==2015))]
```

```python
#Creating new csv dataset by merging the two filtered dataframes
detail = pd.merge(fl_mtch, fl_dlvr, left_on="id", right_on="match_id")
detail.to_csv("details.csv")
```
