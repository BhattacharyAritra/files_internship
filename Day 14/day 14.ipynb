```python
import pandas as pd
```

```python
deliveries=pd.read_csv("E:\\2020 WPL WIDHYA\\Day 14\\deliveries.csv")
matches=pd.read_csv("E:\\2020 WPL WIDHYA\\Day 14\\matches.csv")
```

```python
fl_dlv=deliveries[((deliveries.batting_team=='Kings XI Punjab')&(deliveries.bowling_team=='Sunrisers Hyderabad'))
                 |
                 ((deliveries.batting_team=='Sunrisers Hyderabad')&(deliveries.bowling_team=='Kings XI Punjab'))]
fl_dlv
```

```python
fl_mtc=matches[((matches.team1=='Sunrisers Hyderabad')&(matches.team2=='Kings XI Punjab'))
               |
              ((matches.team1=='Kings XI Punjab')&(matches.team2=='Sunrisers Hyderabad'))]
fl_mtc
```

```python
match_details=pd.merge(left=fl_mtc, right=fl_dlv, left_on='id', right_on='match_id')
match_details.to_csv('plays.csv')
```
