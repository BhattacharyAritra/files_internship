```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
## creating a new subset from the previously existing dataset
subset_dataset = dataset.loc[:, 
[
    'btc_market_cap', 
    'btc_n_transactions', 
    'btc_miners_revenue', 
    'btc_cost_per_transaction', 
    'btc_difficulty', 
    'btc_hash_rate', 
    'btc_cost_per_transaction_percent',
    'btc_market_price' 
]]
subset_dataset.head(10)
```

```python
subset_dataset.isnull().values.any()
```

```python
x = subset_dataset.iloc[:, :-1] ## importing everything except last column
y = subset_dataset.iloc[:, -1]  ## importing only the last column
```

```python
print(x)
```

```python
print(y)
```

```python
## creating train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
## fitting to train model
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_predict = regressor.predict(x_test)
print(y_predict)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_predict, y_test)  ## mean square error value - 425.7471758411076
```
